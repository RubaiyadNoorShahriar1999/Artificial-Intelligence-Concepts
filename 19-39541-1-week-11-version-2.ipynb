```python
cgraph = {'WA': ['SA', 'NT'], 
          'Q': ['SA', 'NT', 'NSW'], 
          'T': [],
          'V': ['SA', 'NSW'], 
          'SA': ['WA', 'NT', 'Q', 'NSW','V'], 
          'NT': ['Q', 'WA', 'SA'], 
          'NSW': ['Q', 'V', 'SA']}
okColor={}
```

```python
# write a method to get the highest degree node

def get_node_hd():
    mnode = ''
    length = 0
    for k in cgraph.keys():
        if(len(cgraph[k])>length and k not in okColor.keys()):
            length = len(cgraph[k])
            mnode = k
    return mnode
```

```python
colorstochooselist={'WA': ['red','green', 'blue'], 
                    'Q': ['red','green', 'blue'],
                    'T': ['red', 'green', 'blue'],
                    'V':  ['red','green', 'blue'], 
                    'NT': ['red','green', 'blue'],
                    'SA': ['red','green', 'blue'],                    
                    'NSW': ['red','green', 'blue']}

import random
def get_color(node):
    # check colorstochooselist[node] is empty or not
    # return 'no color'
    if colorstochooselist[node]!=[]:
        color = random.choice(colorstochooselist[node])
        colorstochooselist.pop(node)
        # update color options
        adjList = cgraph[node]
        for adj in adjList:
            if(adj in colorstochooselist.keys() and color in colorstochooselist[adj]):
                colorstochooselist[adj].remove(color)
        #print(adjList)
        return color
    return "There is No color found"
```

```python
# assign color to the highest degree node
hdNode = get_node_hd()
okColor[hdNode]= get_color(hdNode)
print(colorstochooselist)
print(okColor)
```

```python
# define a method to select a node by MRV
def get_node_MRV():
    #implement the logic list
    node = ''
    length = 8
    for k in colorstochooselist.keys():
        if(len(colorstochooselist[k])<length and k not in okColor.keys()):
            length = len(colorstochooselist[k])
            node = k
    return node
```

```python
L = len(colorstochooselist)
for i in range(0,L):
    # assign color to the Minimum Remaining Value
    mrvNode = get_node_MRV()
    selected_color = get_color(mrvNode)
    print(selected_color)
    # if no color returned then print failure
    if (selected_color=="There is no color found "):
        print("Unsuccessful")
        break
    okColor[mrvNode]= selected_color
    print(colorstochooselist)
    print(okColor)
```
