# def Practice

```python
class Student: 
     
   
    attr1 = "Rubaiyad"
    attr2 = "Munem"
   
    def test(self): 
        print("I'm ", self.attr1)
        print("He is ", self.attr2)

std = Student()
 
print(std.attr1)
std.test()
```

```python
class Food: 
     
   
    attr1 = "Pizza"
    attr2 = "Burger"
   
    def test(self): 
        print("This is ", self.attr1)
        print("This is ", self.attr2)

fd = Food()
 
print(fd.attr1)
print(fd.attr2)
fd.test()
```
