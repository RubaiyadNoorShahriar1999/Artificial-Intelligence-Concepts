```python

import random
def initPopulation(n,b):
    p={}
    for i in range(n):
        s=[]
        for j in range(b):
            s.append(random.choice([0,1]))    
        p[i]=s
    return p      

```

```python
p=initPopulation(4,5)
p
```

```python
def getFitness(p):
    f=[]
    for k in p.keys():
        d = 0
        b = len(p[k])
        for i in range(b-1,-1,-1):
            d=d+p[k][i]*2**(4-i)
        f.append(d**2)
    return f
```

```python
fitness = getFitness(p)
fitness/sum(fitness)
```
