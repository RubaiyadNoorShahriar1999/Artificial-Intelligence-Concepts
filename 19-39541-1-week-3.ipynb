# Table Driven Agent Practice

```python
table={('Loc1','Dry'):'Right',('Loc1','Wet'):'Trun On',('Loc2','Dry'):'Left',('Loc2','Wet'):'Turn On'}

percepts=[]  

def table_driven_agent(percept):
    print(percept)
    percepts.append(percept)
    action = lookup(percept,table)
    return action

def lookup(percept,table):
    for key in table:
        if key == percept:
            return table[key]

import random
Location = random.choice(['Loc1','Loc2'])
Condition = random.choice(['Dry','Wet'])

while True:
    action= table_driven_agent((Location,Condition))
    print(action)
    cmd = input()
    if(cmd == 'stop'): break
    if action == 'Right':
        Location = 'Loc2'
        Condition = random.choice(['Dry','Wet'])
    elif action== 'Left':
        Location = 'Loc1'
        Condition = random.choice(['Dry','Wet'])
    else:
        Condition = 'Dry'   
```

# Simple Reflex Agent Practice 

```python
def simple_reflex_agent(percept):
    print(percept)
    if percept[1]=='Wet':
        action = 'Turn On'
    elif percept[0]=='Loc1':
        action = 'Right'
    elif percept[0]=='Loc2':
        action = 'Left'
    return action

import random
Location = random.choice(['Loc1','Loc2'])
Condition = random.choice(['Dry','Wet'])

while True:
    action= simple_reflex_agent((Location,Condition))
    print(action)
    cmd = input()
    if(cmd == 'stop'): break
    if action == 'Right':
        Location = 'Loc2'
        Condition = random.choice(['Dry','Wet'])
    elif action== 'Left':
        Location = 'Loc1'
        Condition = random.choice(['Dry','Wet'])
    else:
        Condition = 'Dry'

    
```

# Model Based Agent Practice 

```python
model = {'Loc1':'Unknown', 'Loc2':'Unknown'}
state = 'WET'
action = 'Unknown'

def update_state(percept, model):
    global state, action
    if action=='Turn On':
        model[percept[0]]='Dry'
    if model['Loc1']=='Dry' and model['Loc2']=='Dry':
        state = 'DRY'
    else:
        state = 'WET'
    return state
        

def model_based_reflex_agent(percept):
    global state,action,model
    model[percept[0]]=percept[1]
    state = update_state(percept, model)
    print('Perception: '+str(percept)) 
    print('Model: '+str(model))   
    print('State: '+str(state))  
    if state == 'DRY':
        action = 'Stand Still'
    elif percept[1]=='Wet':
        action = 'Turn On'
    elif percept[0]=='Loc1':
        action = 'Right'
    elif percept[0]=='Loc2':
        action = 'Left'
    return action


import random
Location = random.choice(['Loc1','Loc2'])
Condition= random.choice(['Dry','Wet'])

while True:
    action = model_based_reflex_agent((Location,Condition))
    print('Action: '+ str(action))
    if action == 'Right':
        Location = 'Loc2'
        Condition = random.choice(['Dry','Wet'])
    elif action == 'Left':
        Location = 'Loc1'
        Condition = random.choice(['Dry','Wet'])
    elif action == 'Turn On':
        Condition = 'Dry'
    elif action == 'Stand Still':
        break
        

    
```
