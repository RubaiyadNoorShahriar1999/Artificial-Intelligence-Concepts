```python
import random
def initPopulation(n,b):
    p={}
    for i in range(n):
        s=[]
        for j in range(b):
            s.append(random.choice([0,1]))    
        p[i]=s
    return p      

```

```python
population=initPopulation(8,5)
population
```

# Step 2: Parent Selection

Fitness Calculation

```python
[i for i in range(4,-1,-1)]
```

```python
def getFitness(p):
    f=[]
    for k in p.keys():
        d = 0
        b = len(p[k])
        for i in range(b-1,-1,-1):
            d=d+p[k][i]*2**(4-i)
        f.append(d**2)
    return f
```

```python
fitness = getFitness(population)
fitness
```

probability

```python
total_fitness = sum(fitness)
prob = [round(f/total_fitness,3) for f in fitness]
prob
```

```python
def select_parent(n,probabilities):
    parents = {}
    for i in range(n):
        r =random.choice(range(1,1001,1))/1000
        s = 0
        for j in range(len(probabilities)):
            s = s + probabilities[j]
            if(s>r):
                parents[i]=population[j]
                #print(population[j])
                break
    return parents
```

```python
parents = select_parent(6,prob)
parents
```

# Step 4: Survivor (will go to population)

Select top 2 offspring based on fitness value and add them to population.

```python
def crossover(parents,cp):
    offs = []
    for i in parents.keys():
        
        if i%2==0:
            os1 = parents[i][:cp] + parents[i+1][cp:]
            os2 = parents[i+1][:cp] + parents[i][cp:]
            offs.append(os1)
            offs.append(os2)
            
    offsprings = {i:offs[i] for i in range(len(offs))}
    
    return offsprings
```
