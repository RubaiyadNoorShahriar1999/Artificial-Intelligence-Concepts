# An example of method

```python
def get_names(name):
    name_list = name.split()
    fn = name_list[0]
    ln = name_list[-1]
    mn = ''
    if len(name_list)>2:
        mn = name_list[1]
        for v in name_list[2:-1]:
            mn = mn + ' ' +v
    return fn,mn,ln
```

# Table driven agent

```python
table={('A','Clean'):'Right',('A','Dirty'):'Suck',('B','Clean'):'Left',('B','Dirty'):'Suck'}
percepts=[]    
def table_driven_agent(percept):
    print(percept)
    percepts.append(percept)
    action = lookup(percept,table)
    return action

def lookup(percept,table):
    for key in table:
        if key == percept:
            return table[key]

import random
Location = random.choice(['A','B'])
Condition = random.choice(['Clean','Dirty'])

while True:
    action= table_driven_agent((Location,Condition))
    print(action)
    cmd = input()
    if(cmd == 'stop'): break
    if action == 'Right':
        Location = 'B'
        Condition = random.choice(['Clean','Dirty'])
    elif action== 'Left':
        Location = 'A'
        Condition = random.choice(['Clean','Dirty'])
    else:
        Condition = 'Clean'   
```

# Simple reflex agent

```python
def simple_reflex_agent(percept):
    print(percept)
    if percept[1]=='Dirty':
        action = 'Suck'
    elif percept[0]=='A':
        action = 'Right'
    elif percept[0]=='B':
        action = 'Left'
    return action

import random
Location = random.choice(['A','B'])
Condition = random.choice(['Clean','Dirty'])

while True:
    action= simple_reflex_agent((Location,Condition))
    print(action)
    cmd = input()
    if(cmd == 'stop'): break
    if action == 'Right':
        Location = 'B'
        Condition = random.choice(['Clean','Dirty'])
    elif action== 'Left':
        Location = 'A'
        Condition = random.choice(['Clean','Dirty'])
    else:
        Condition = 'Clean'

    
```

# Model based agent

```python
model = {'A':'Unknown', 'B':'Unknown'}
state = 'Bad'
action = 'Unknown'

def update_state(percept, model):
    global state, action
    if action=='Suck':
        model[percept[0]]='Clean'
    if model['A']=='Clean' and model['B']=='Clean':
        state = 'Good'
    else:
        state = 'Bad'
    return state
        

def model_based_reflex_agent(percept):
    global state,action,model
    model[percept[0]]=percept[1]
    state = update_state(percept, model)
    print('Perception: '+str(percept)) 
    print('Model: '+str(model))   
    print('State: '+str(state))  
    if state == 'Good':
        action = 'Nothing to do'
    elif percept[1]=='Dirty':
        action = 'Suck'
    elif percept[0]=='A':
        action = 'Right'
    elif percept[0]=='B':
        action = 'Left'
    return action


import random
Location = random.choice(['A','B'])
Condition= random.choice(['Clean','Dirty'])

while True:
    action = model_based_reflex_agent((Location,Condition))
    print('Action: '+ str(action))
    if action == 'Right':
        Location = 'B'
        Condition = random.choice(['Clean','Dirty'])
    elif action == 'Left':
        Location = 'A'
        Condition = random.choice(['Clean','Dirty'])
    elif action == 'Suck':
        Condition = 'Clean'
    elif action == 'Nothing to do':
        break
        

    
```
