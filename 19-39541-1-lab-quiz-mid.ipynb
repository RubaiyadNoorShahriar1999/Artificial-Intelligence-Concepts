# Table Driven Agent

```python
def table_driven_agent(percept): # percept contains the present state of the vaccum cleaner
    
    percepts=[]                  # percepts contains the history of previous percept.
    table=[['dirty','suck'],['clean','move']]
    percepts.append(percept)
    action = lookup(percept,table) # Action contains what to do.
    
    if action == 'move':
       action = move(percept[1])
    return(action)

def lookup(percept,table): # Lookup() will help in finding the action for the current status.
    
    for x in table:
        if x[0] == percept[0]:
            return(x[1])
        
def move(percept): # move() function helps the vaccum if the current status is clean to which way to move.
    
    if percept == 'A':
        return('Go to B')
    elif percept == 'B':
        return('Go to C')
    elif percept == 'C':
        return('Go to A')
    
percept = list(map(str,input("Enter the status and postion: ").split(',')))
action = table_driven_agent(percept)
print(action)
```
