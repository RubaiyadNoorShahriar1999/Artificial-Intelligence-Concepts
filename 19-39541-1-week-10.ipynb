# N Queen Attack Detection

```python
import random
N = 4
r = []
c = []
for i in range(1, N+1):
    r1 = random.choice(range(1, N+1))
    c1 = random.choice(range(1, N+1))
    while (r1 in r) and (c1 in c) and (r.index(r1)==c.index(c1)):
        r1 = random.choice(range(1, N+1))
        c1 = random.choice(range(1, N+1))
    r.append(r1)
    c.append(c1)
    print('Q{}: {},{}'.format(i,r[-1],c[-1]))
print(r)
print(c)
```

```python
# row attack :
for i in range(0,N):
    for j in range(i+1,N):
        if(r[i] == r[j]):
            print('row attack: row:{} column:{} and {}'.format(r[i],c[i],c[j]))      
    
```

```python
# column attack :
for i in range(0,N):
    for j in range(i+1,N):
        if(c[i]==c[j]):
            print('col attack: col:{} row:{} and {}'.format(c[i],r[i],r[j]))
```

```python
# diagonal atack :
for i in range(0,N):
    for j in range(i+1,N):
        if(abs(r[i]-r[j])==abs(c[i]-c[j])):
            print('diagonal attack',(r[i],c[i]))
            print(r[j],c[j])
```

```python
def nqAttack(r,c):
    for i in range(0,N):
        for j in range(i+1,N):
            if(r[i] == r[j]):
                print('Row attack: row: {}  column: {} and {}'.format(r[i],c[i],c[j]))
            if(c[i] == c[j]):
                print('Column attack: column: {}  row: {} and {}'.format(c[i],r[i],r[j]))
            if(abs(r[i]-r[j])==abs(c[i]-c[j])):
                print('Diagonal attack: position: {},{} and {},{}'.format(r[i],c[i],r[j],c[j]))
```

```python
nqAttack(r,c)
```
