```python
romanian_map = {
    'Arad': {'Sibiu':140, 'Zerind':75, 'Timisoara':118},
    'Zerind': {'Arad':75, 'Oradea':71},
    'Oradea': {'Zerind':71, 'Sibiu': 151},
    'Sibiu': {'Arad':140, 'Oradea':151, 'Fagaras':99, 'Rimnicu':80},
    'Timisoara': {'Arad': 118, 'Lugoj':111},
    'Lugoj': {'Timisoara':111, 'Mehadia':70},
    'Mehadia': {'Lugoj': 70, 'Drobeta':75},
    'Drobeta': {'Mehadia':75, 'Craiova':120},
    'Craiova': {'Drobeta': 120, 'Rimnicu': 146, 'Pitesti': 138},
    'Rimnicu': {'Sibiu': 80, 'Craiova': 146, 'Pitesti': 97},
    'Fagaras': {'Sibiu':99, 'Bucharest':211},
    'Pitesti': {'Rimnicu': 97, 'Craiova':138, 'Bucharest':101},
    'Bucharest': {'Fagaras':211, 'Pitesti': 101, 'Giurgiu':90, 'Urziceni':85},
    'Giurgiu': {'Bucharest': 90},
    'Urziceni': {'Bucharest': 85, 'Vaslui':142, 'Hirsova':98},
    'Hirsova': {'Urziceni':98, 'Eforie':86},
    'Eforie': {'Hirsova':86},
    'Vaslui': {'Iasi': 92, 'Urziceni':142},
    'Iasi': {'Vaslui':92, 'Neamt': 87},
    'Neamt': {'Iasi':87}
}
```

```python
import collections
from queue import Queue


def bfs(startingNode, destinationNode):
    visited = {}#to store explored nodes
    distance = {}#will print the toatl distance
    parent = {}#to store path
    traversList = []
    queue = collections.deque([startingNode])
    traversList.append(startingNode)


    bfs_traversal_output = []#to store the order of traversal
    frontier = Queue()#initializing queue

    for city in romanian_map.keys():
        visited[city] = False #As we did not define the starting node
        parent[city] = None #As we did not start traversing
        distance[city] = -1 #For initial state

    startingCity = startingNode #Assigning the starting node to the startingCity valriable
    visited[startingCity] = True #As startingCity is initialized the starting node is visited 
    distance[startingCity] = 0 #Defines the starting node
    frontier.put(startingCity)    # Pushing the startingNode value to the queue(Arad)
    found = False #initially false for the incoming while loop to be executed 

    while not frontier.empty(): #If the frontier is not empty then this while loop will be executed 
        u = frontier.get()     #In the first iteration it will be arad
        bfs_traversal_output.append(u) #Keeps track of the path the system will traverse and get the solution
        vertex = queue.popleft() 
        for v in romanian_map[u].keys(): #v is the value of visited node and in romanian_map[u].keys(), u is the parent node and it will print the children of the parent node u from which v will be searched 
            if not visited[v]:  #if v is not visted then it will run              
                visited[v] = True
                parent[v] = u
                distance[v] = distance[u] + romanian_map[u][v] #It will append the parent nodes distance fron strtig node and distance between parent node u and child node v            
                frontier.put(v) # Will push the visited node in the queue
                traversList.append(v)
                queue.append(v)
                if v== destinationNode: #Here it will check if the vsited node is the destination or not
                    found =True
                    break #Will vreak form the for loop
                
        if found: break #If the destination if found it will break from the while loop from here
            
    g = destinationNode 
    path = []
    while g is not None:
        path.append(g)
        g = parent[g]
    
    path.reverse() #reversing the list as it was appended in the oposite direction from the destinationNode
    print(path) #the final path
    print(traversList) #traversing path list
    print(distance[destinationNode]) #resultant distance
    print(bfs_traversal_output)#BFS traversal list

bfs('Arad', 'Bucharest')
```

# Uniform Cost Search

```python
from queue import PriorityQueue

def ucs(startingNode, destinationNode):
    visited = {} #to store explored nodes
    distance = {} #will print the toatl distance
    parent = {} #to store path
    traversList = []
    traversList.append(startingNode)

    ucs_traversal_output = [] #to store the order of traversal
    pq = PriorityQueue() #we will decide about the next node according to the priority of the minimum cost 

    for city in romanian_map.keys():
        visited[city] = False #As we did not define the starting node
        parent[city] = None  #As we did not start traversing
        distance[city] = -1 #For initial state

    startingCity = startingNode #Assigning the starting node to the startingCity valriable
    visited[startingCity] = True #As startingCity is initialized the starting node is visited
    distance[startingCity] = 0 #Defines the starting node
    pq.put((0,startingCity)) #pushing the distance and the city name to the Priority queue.Distance value is needed to find the shortest path 

    while not pq.empty(): #will run if and only if the pq queue isn't empty
        u = pq.get()[1]  #will return distance with minimum cost and city name    
        ucs_traversal_output.append(u) #will add the visited u node to the traversal output list
        if u== destinationNode: #if we find the destination node which is u,the loop will break
            break
        visited[u] = True #for every iteraration the visited node u will be marked as visited
        for v in romanian_map[u].keys(): #v is the value of visited node and in romanian_map[u].keys(), u is the parent node and it will print the children of the parent node u from which v will be searched 
            if not visited[v]:                
                parent[v] = u
                distance[v] = distance[u] + romanian_map[u][v] #to find the shortest path
                pq.put((distance[v],v)) #will push the distance and the city name
                traversList.append(v)

    g = destinationNode
    path = []
    while g is not None:
        path.append(g)
        g = parent[g]

    path.reverse()  #reversing the list as it was appended in the oposite direction from the destinationNode
    print(path) #the final path
    print(traversList) #traversing path list
    print(distance[destinationNode])#resultant distance
    print(ucs_traversal_output) #UCS traversal list


ucs('Arad', 'Bucharest')
```
